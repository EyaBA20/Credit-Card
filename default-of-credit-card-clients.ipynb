```python
#rendre toutes les colonnes de type numérique 
for i in df1.columns :
    df1[[i]] = df1[[i]].apply(pd.to_numeric)
```

```python
df1.dtypes
```

```python
#visualiser le boxplot pour cerner les outliers dans la colonne X1
df1[["X1"]].boxplot()
```

```python
#enlever les outliers de la colonne X1
#for i in df1.X1.unique():
    q1 = df1['X1'].quantile(0.25)
    q3 = df1['X1'].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    for j in df1.index:
        if((df1.loc[j,'X1']>fence_high) or (df1.loc[j,'X1']<fence_low)):
            indice=j
            df1.drop(indice, inplace=True)
            
```

```python
#visualiser le boxplot pour s'assurer de la suppression des outliers de la colonne X1
df1[["X1"]].boxplot()
```

```python
#nettoyage de la colonne X3
pd.unique(df1.X3)
```

```python
#cleaning de la colonne X3 => remplacer 0.5.6 par 4
for j in df1.index:
    if (df1.loc[j,'X3']== 5) or (df1.loc[j,'X3']==0) or (df1.loc[j,'X3']==6 ):
        indice=j
        df1.loc[indice,'X3']=4
            
```
